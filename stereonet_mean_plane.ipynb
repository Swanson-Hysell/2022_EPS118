## Plot dikes measurements

We can now plot all the planes on a stereonet.

````python
strikes = mafic_dikes[' strike']
dips = mafic_dikes[' dip']

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='stereonet')

ax.plane(strikes, dips, color='grey', linewidth=2)
ax.pole(strikes, dips, color='red', markersize=8, alpha=0.5)
ax.grid()
fig.tight_layout()

plt.show()
````

## Filter data by location

Often it makes sense to group data by subregion. The code below splits the data based on latitude

````python
north_planes = mafic_dikes[(mafic_dikes[' latitude'] > 37.66)]
south_planes = mafic_dikes[(mafic_dikes[' latitude'] < 37.66)]
````

## We can plot data by subgroup

````python
southern_strikes = south_planes[' strike']
southern_dips = south_planes[' dip']

northern_strikes = north_planes[' strike']
northern_dips = north_planes[' dip']

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='stereonet')

ax.plane(southern_strikes, southern_dips, color='red', linewidth=2, alpha=0.5)
ax.pole(southern_strikes, southern_dips, color='red', markersize=8, alpha=0.5)

ax.plane(northern_strikes, northern_dips, color='blue', linewidth=2, alpha=0.5)
ax.pole(northern_strikes, northern_dips, color='blue', markersize=8, alpha=0.5)

ax.grid()
fig.tight_layout()
plt.show()
````

## Find the average strike dip

Using the function ```mplstereonet.fit_pole()``` the mean plane can be determined from a set of strike dip measurements. The code below calculates the mean plane and then plots it along with the individual plane measurements.

````python
mean_s_mafic_strike, mean_s_mafic_dip = mplstereonet.fit_pole(southern_strikes, southern_dips)
print('The mean strike is:')
print(round(mean_s_mafic_strike,1))
print('The mean dip is:')
print(round(mean_s_mafic_dip,1))
````

````python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='stereonet')

ax.plane(southern_strikes, southern_dips, color='pink', linewidth=2, alpha=0.5)
ax.pole(southern_strikes, southern_dips, color='pink', markersize=8, alpha=0.5)

ax.plane(mean_s_mafic_strike, mean_s_mafic_dip, color='red', linewidth=5, alpha=1,label='mean plane')
ax.pole(mean_s_mafic_strike, mean_s_mafic_dip, color='red', markersize=12, alpha=1)

plt.legend()
ax.grid()
fig.tight_layout()
plt.show()
````
